# Merge

```python
import pandas as pd
```

```python
var1 = pd.DataFrame({"A": [1,3,4,5], "B" : [11,22,33,44]})
var1
```

```python
var2 = pd.DataFrame({"A": [1,3,4,5], "C" : [23,34,56,66]})
var2
```

```python
var1 = pd.DataFrame({"A": [1,3,4,5], "B" : [11,22,33,44]})
var2 = pd.DataFrame({"A": [1,3,4,5], "C" : [23,34,56,66]})

pd.merge(var1,var2, on = "A")
```

```python
var1 = pd.DataFrame({"A": [1,3,4,5], "B" : [11,22,33,44]})
var2 = pd.DataFrame({"A": [1,3,4,5], "C" : [23,34,56,66]})

pd.merge(var2,var1, on = "A")
```

```python
var3 = pd.DataFrame({"A": [1,3,4,5], "B" : [11,22,33,44]})
var4 = pd.DataFrame({"A": [1,2,4,5], "C" : [23,34,56,66]})

pd.merge(var3,var4, how = "inner")
```

```python
var5 = pd.DataFrame({"A": [1,3,4,5], "B" : [11,22,33,44]})
var6 = pd.DataFrame({"A": [1,2,4,5], "C" : [23,34,56,66]})

pd.merge(var5,var6, how = "outer")
```

```python
var7 = pd.DataFrame({"A": [1,3,4,5], "B" : [11,22,33,44]})
var8 = pd.DataFrame({"A": [1,2,4,5], "C" : [23,34,56,66]})

pd.merge(var7,var8, how = "outer",indicator = "T")
```

# Concat

```python
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([11,21,31,41])

pd.concat([sr1,sr2])
```

```python
d1 = pd.DataFrame({"A":[1,3,4,5],"B":[11,22,44,55]})
d2 = pd.DataFrame({"A":[1,3,4,5],"B":[21,23,24,26]})

pd.concat([d1,d2])
```

```python
d3 = pd.DataFrame({"A":[1,3,4,5],"B":[11,22,44,55]})
d4 = pd.DataFrame({"A":[1,3,4,5],"B":[21,23,24,26]})

pd.concat([d3,d4],axis=1,keys=["Python","Java"]) // According to columns
# print()
# pd.concat([d3,d4],axis=0,keys=["Python","Java"]) //  According to rows
```
