# Subplot

Matplotlib Subplot:

- With the subplot() function we can draw multiple plots in one figure.

```python
# Drow two plot
# plot 1

x = np.array([0, 1, 3, 4])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x, y)

# Plot 2

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)

plt.plot(x, y)

plt.show()
```

subplot() Function:

- The subplot() function takes three arguments that describes the layout of the figure.

- The layout is organized in rows and columns, which are represented by the first and second argument.

- The third argument represents the index of the current plot.

- Syntax:

- plt.subplot(rows, cols, index)


- rows: Number of rows for organization of subplots.

- cols: Number of columns for organization of subplots.

- index: Index number of the current subplot, starts from 1.

```python
# Drow the 6 plot
# plot 1
x = np.array([0, 1, 2, 3])
y = np.array([15, 20, 30, 40])

plt.subplot(2, 3, 1)
plt.plot(x, y)
# plot 2
x = np.array([0, 1, 2, 3])
y = np.array([30, 10, 70, 80])

plt.subplot(2, 3, 2)
plt.plot(x, y)
# plot 3
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x, y)
# plot 4
x = np.array([0, 1, 2, 3])
y = np.array([10, 2, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x, y)
# plot 5
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x, y)
# plot 6
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 15, 40])

plt.subplot(2, 3, 6)
plt.plot(x, y)

plt.show()
```

- Title each subplot:

- We can add a title to each plot with the title() function.

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Match")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([15, 20, 30, 45])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Summary")

plt.show()
```

- Super Title:

- We can add a title to the entire figure with the suptitle() function.

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Match")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([15, 20, 30, 45])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Summary")

plt.suptitle("IND VS ASU")

plt.show()
```

```python
x = np.linspace(1, 15, 10)
y = x**2
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'b--')
plt.subplot(1,2,2)
plt.plot(x, y, 'r-*')

plt.show()
```

```python
# plot the subgraph 2 by 2 matrix

x = np.linspace(1, 15, 10)
y = x**2

plt.subplot(2,2,1)
plt.plot(x, y, 'b--')
plt.subplot(2,2,2)
plt.plot(x, y, 'r-*')
plt.subplot(2,2,3)
plt.plot(x, y, 'g-H')
plt.subplot(2,2,4)
plt.plot(x, y, 'y>')

plt.show()
```

Figure function:

```python
x = np.linspace(1, 15, 10)
y = x**2

# Graph in side the graph.......

fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

ax1.plot(x, y)
ax2.plot(y, x)

plt.show()
```
