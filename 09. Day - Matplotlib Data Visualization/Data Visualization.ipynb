# Subplot

```python
x = np.linspace(1, 15, 10)
y = x**2
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'b--')
plt.subplot(1,2,2)
plt.plot(x, y, 'r-*')

plt.show()
```

```python
# plot the subgraph 2 by 2 matrix

x = np.linspace(1, 15, 10)
y = x**2

plt.subplot(2,2,1)
plt.plot(x, y, 'b--')
plt.subplot(2,2,2)
plt.plot(x, y, 'r-*')
plt.subplot(2,2,3)
plt.plot(x, y, 'g-H')
plt.subplot(2,2,4)
plt.plot(x, y, 'y>')

plt.show()
```

Figure function:

```python
x = np.linspace(1, 15, 10)
y = x**2

# Graph in side the graph.......

fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

ax1.plot(x, y)
ax2.plot(y, x)

plt.show()
```
